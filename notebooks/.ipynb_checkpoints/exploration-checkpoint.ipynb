## 1️⃣ Setup & Imports

```python
import os
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
import cv2
import torch
from PIL import Image
from collections import Counter
from pathlib import Path

sys.path.append('../src')
from utils import set_seed, get_device
from preprocessing import get_dataloaders, load_image_paths_and_labels

set_seed(42)
device = get_device()

# Dataset path
DATA_ROOT = '../data/chest_xray'
print(f'Dataset root: {DATA_ROOT}')
print(f'Device: {device}')
```

---
## 2️⃣ Dataset Statistics

```python
# Count images per split and class
splits = ['train', 'val', 'test']
classes = ['NORMAL', 'PNEUMONIA']
stats = {}

for split in splits:
    stats[split] = {}
    for cls in classes:
        path = os.path.join(DATA_ROOT, split, cls)
        count = len(os.listdir(path)) if os.path.exists(path) else 0
        stats[split][cls] = count

df_stats = pd.DataFrame(stats).T
df_stats['Total'] = df_stats.sum(axis=1)
df_stats['Imbalance Ratio'] = (df_stats['PNEUMONIA'] / df_stats['NORMAL']).round(2)
print('\n📊 Dataset Statistics:')
print(df_stats)
```

```python
# Visualize class distribution
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
colors = ['#00c851', '#ff4444']

for ax, split in zip(axes, splits):
    values = [stats[split]['NORMAL'], stats[split]['PNEUMONIA']]
    bars = ax.bar(classes, values, color=colors, edgecolor='white', width=0.5)
    ax.set_title(f'{split.upper()} Set', fontweight='bold', fontsize=13)
    ax.set_ylabel('Number of Images')
    for bar, val in zip(bars, values):
        ax.text(
            bar.get_x() + bar.get_width()/2,
            bar.get_height() + 10,
            str(val), ha='center', fontweight='bold'
        )
    ax.grid(axis='y', alpha=0.3)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

plt.suptitle('Class Distribution Across Splits',
             fontsize=15, fontweight='bold')
plt.tight_layout()
plt.savefig('../outputs/eda_class_distribution.png', dpi=150)
plt.show()
```

---
## 3️⃣ Sample Images Visualization

```python
# Show sample images from each class
fig, axes = plt.subplots(2, 5, figsize=(20, 8))

for row, cls in enumerate(classes):
    cls_path = os.path.join(DATA_ROOT, 'train', cls)
    images   = os.listdir(cls_path)[:5]
    for col, img_name in enumerate(images):
        img = Image.open(os.path.join(cls_path, img_name)).convert('RGB')
        axes[row][col].imshow(img, cmap='gray')
        axes[row][col].set_title(
            f'{cls}\n{img.size[0]}x{img.size[1]}',
            fontsize=9,
            color='green' if cls == 'NORMAL' else 'red',
            fontweight='bold'
        )
        axes[row][col].axis('off')

plt.suptitle('Sample Chest X-Ray Images',
             fontsize=15, fontweight='bold')
plt.tight_layout()
plt.savefig('../outputs/eda_sample_images.png', dpi=150)
plt.show()
```

---
## 4️⃣ Image Properties Analysis

```python
# Analyze image dimensions and pixel statistics
widths, heights, means, stds = [], [], [], []
sample_size = 200

for cls in classes:
    cls_path = os.path.join(DATA_ROOT, 'train', cls)
    images   = os.listdir(cls_path)[:sample_size//2]
    for img_name in images:
        img = cv2.imread(os.path.join(cls_path, img_name), cv2.IMREAD_GRAYSCALE)
        if img is not None:
            widths.append(img.shape[1])
            heights.append(img.shape[0])
            means.append(img.mean())
            stds.append(img.std())

print(f'📐 Image Dimensions:')
print(f'   Width  → Min: {min(widths)} | Max: {max(widths)} | Mean: {np.mean(widths):.0f}')
print(f'   Height → Min: {min(heights)} | Max: {max(heights)} | Mean: {np.mean(heights):.0f}')
print(f'\n🎨 Pixel Statistics (grayscale):')
print(f'   Mean Intensity → {np.mean(means):.2f}')
print(f'   Mean Std       → {np.mean(stds):.2f}')
```

```python
# Pixel intensity distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

for cls, color in zip(classes, ['#00c851', '#ff4444']):
    cls_path  = os.path.join(DATA_ROOT, 'train', cls)
    all_pixels = []
    for img_name in os.listdir(cls_path)[:50]:
        img = cv2.imread(
            os.path.join(cls_path, img_name),
            cv2.IMREAD_GRAYSCALE
        )
        if img is not None:
            img = cv2.resize(img, (224, 224))
            all_pixels.extend(img.flatten().tolist())

    axes[0].hist(all_pixels, bins=50, alpha=0.6,
                 color=color, label=cls, density=True)

axes[0].set_title('Pixel Intensity Distribution',
                   fontweight='bold')
axes[0].set_xlabel('Pixel Value')
axes[0].set_ylabel('Density')
axes[0].legend()
axes[0].grid(alpha=0.3)

# Dimension scatter
axes[1].scatter(widths, heights, alpha=0.5, color='#00d4ff', s=20)
axes[1].set_title('Image Dimensions Scatter',
                   fontweight='bold')
axes[1].set_xlabel('Width (px)')
axes[1].set_ylabel('Height (px)')
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.savefig('../outputs/eda_pixel_analysis.png', dpi=150)
plt.show()
```

---
## 5️⃣ CLAHE Enhancement Visualization
*(Shows why CLAHE preprocessing helps detect early pneumonia patterns)*

```python
# Compare original vs CLAHE enhanced images
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))

fig, axes = plt.subplots(2, 4, figsize=(20, 10))

for row, cls in enumerate(classes):
    cls_path = os.path.join(DATA_ROOT, 'train', cls)
    images   = os.listdir(cls_path)[:2]

    for col_offset, img_name in enumerate(images):
        img = cv2.imread(
            os.path.join(cls_path, img_name),
            cv2.IMREAD_GRAYSCALE
        )
        img         = cv2.resize(img, (224, 224))
        img_clahe   = clahe.apply(img)
        col_orig    = col_offset * 2
        col_clahe   = col_offset * 2 + 1

        axes[row][col_orig].imshow(img, cmap='gray')
        axes[row][col_orig].set_title(
            f'{cls} - Original',
            color='green' if cls == 'NORMAL' else 'red',
            fontweight='bold'
        )
        axes[row][col_orig].axis('off')

        axes[row][col_clahe].imshow(img_clahe, cmap='gray')
        axes[row][col_clahe].set_title(
            f'{cls} - CLAHE Enhanced',
            color='green' if cls == 'NORMAL' else 'red',
            fontweight='bold'
        )
        axes[row][col_clahe].axis('off')

plt.suptitle(
    'CLAHE Enhancement — Revealing Early Pneumonia Patterns',
    fontsize=14, fontweight='bold'
)
plt.tight_layout()
plt.savefig('../outputs/eda_clahe_comparison.png', dpi=150)
plt.show()
```

## 6️⃣ DataLoader Verification

```python
# Verify DataLoaders work correctly
dataloaders, class_counts = get_dataloaders(DATA_ROOT, batch_size=8)

# Get one batch
images, labels = next(iter(dataloaders['train']))

print(f'✅ Batch shape:  {images.shape}')
print(f'✅ Labels shape: {labels.shape}')
print(f'✅ Label values: {labels.tolist()}')
print(f'✅ Image dtype:  {images.dtype}')
print(f'✅ Image range:  [{images.min():.3f}, {images.max():.3f}]')
```

```python
# Visualize augmented batch
mean = np.array([0.485, 0.456, 0.406])
std  = np.array([0.229, 0.224, 0.225])

fig, axes = plt.subplots(2, 4, figsize=(18, 9))
class_names = ['NORMAL', 'PNEUMONIA']
colors      = ['#00c851', '#ff4444']

for idx, ax in enumerate(axes.flatten()):
    if idx >= len(images):
        break
    img = images[idx].numpy().transpose(1, 2, 0)
    img = np.clip(std * img + mean, 0, 1)
    ax.imshow(img)
    label = labels[idx].item()
    ax.set_title(
        class_names[label],
        color=colors[label],
        fontweight='bold'
    )
    ax.axis('off')

plt.suptitle(
    'Augmented Training Batch (after Albumentations)',
    fontsize=14, fontweight='bold'
)
plt.tight_layout()
plt.savefig('../outputs/eda_augmented_batch.png', dpi=150)
plt.show()
```
